```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Cutlets.csv')
df
```

```python
sns.distplot(df['Unit A'],hist = False)
sns.distplot(df['Unit B'],hist = False)
plt.legend(['Unit A','Unit B'])
plt.show()
```

```python
stats.ttest_ind(df['Unit A'],df['Unit B'])
```

compare p_value with α = 0.05 (At 5% significance level)

**Null Hypothesis Accepted(Pvalue > 0.05) and ther is no significant diffrence between Unit A and Unit B**
