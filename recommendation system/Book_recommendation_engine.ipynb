### Use cosin as a metric for UBCF

```python
books_pivot_table = pd.pivot_table(data = books_data, values='Book_rating',index='User_id',columns='Book_title').fillna(0)
books_pivot_table.index = books_data.User_id.unique()
books_pivot_table
```

```python
values = books_pivot_table.values
values
```

```python
cosine_ubcf = 1 - pairwise_distances(X = values, metric='cosine')
cosine_ubcf
```

```python
cosine_ubcf_df = pd.DataFrame(data = cosine_ubcf)
cosine_ubcf_df.index = books_data.User_id.unique()
cosine_ubcf_df.columns = books_data.User_id.unique()
cosine_ubcf_df
```

**FILLING THE DIAGONAL WITH ZERO**

```python
np.fill_diagonal(cosine_ubcf_df.to_numpy(),val = 0)
```
